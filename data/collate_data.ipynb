```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

businesses = pd.read_csv("new_business_data.csv", index_col="abn")
businesses
```

```python
contracts = pd.read_csv("contract_data.csv")
# reformat contract data
contracts["value"] = contracts["value"].apply(lambda x: float(x.replace("$", "").replace(",", "")))
contracts["start"] = pd.to_datetime(contracts["period"].apply(lambda x: x.split(" to ")[0]), format="%d-%b-%Y")
contracts["end"] = pd.to_datetime(contracts["period"].apply(lambda x: x.split(" to ")[1]), format="%d-%b-%Y")
contracts["days"] = contracts["end"] - contracts["start"]
contracts = contracts[(contracts["start"] > "2015-06-30") & (contracts["start"] < "2020-07-01")]
contracts.sort_values("cn")
```

```python
contract_groups_value = contracts.groupby("abn").sum().sort_values("value", ascending=False)
contract_total = contract_groups_value.sum()[0]
i = 1
shares = []
while True:
    shares.append(contract_groups_value.iloc[:i]["value"].sum() / contract_total)
    if i == len(contract_groups_value):
        break
    i = i + 1
contract_groups_value["shares"] = shares
entities = []
biz_names = []
for abn in contract_groups_value.index:
    entities.append(businesses.loc[abn, "entity"])
    biz_names.append(businesses.loc[abn, "biz_name"])    
contract_groups_value["entity"] = entities
contract_groups_value["biz_name"] = biz_names
contract_groups_value["value"] = contract_groups_value["value"].apply(lambda x: f"${x: ,.0f}")
contract_groups_value[["biz_name", "value"]].head(20)
contract_groups_value
```
